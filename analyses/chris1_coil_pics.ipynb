# Pictures for rapid identification

```python
import numpy as np
import numpy.linalg as nla
import matplotlib.pyplot as plt
import nibabel as nib
import os
import time
```

```python
path = '/home/snarles/predator'
```

```python
vmin_ = -2.0
vmax_ = 17.0
figsize_ = (3, 3)
z_inds = [10, 20, 30, 40, 50]
b0_inds = [0, 1, 16, 31, 46, 61, 75, 90, 105, 120, 135, 149]
```

## Loading

```python
t0 = time.time()
fs = list()
figsize_2 = (len(z_inds) * figsize_[0], figsize_[1])
coil = nib.load(path + '/8631_5_1_pfile/coil_images/coil_comb_ec.nii.gz').get_data()
coil = coil[:, :, :, b0_inds]
coil_mu = np.mean(coil, axis = 3)
f, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(coil_mu[:, :, z_ind], cmap='bone')
fs.append(f)
for i in range(1,32):
    coil = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil' + str(i) + '_ec.nii.gz').get_data()
    coil = coil[:, :, :, b0_inds]
    coil_mu = np.mean(coil, axis = 3)
    f, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
    for j in range(len(z_inds)):
        z_ind = z_inds[j]
        axarr[j].imshow(coil_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
    fs.append(f)
time.time() - t0
```
