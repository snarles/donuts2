```python
vthres = 0.1
def varfilt(st):
    temp = np.array([float(v) for v in str(st).split(' ')])[3:]
    return np.var(temp) > vthres
```

```python
filt_rdd = rawrdd.filter(varfilt)
```

```python
filt_rdd.count()
```

```python
def par(st):
    #temp = np.sqrt(np.array([float(v) for v in str(st).split(' ')])[3:])
    temp = np.array([float(v) for v in str(st).split(' ')])[3:]    
    return standardize(temp)
```

```python
no_rdd = filt_rdd.map(par)
```

```python
model = KMeans.train(no_rdd, 5)
```

```python
def clas(st):
    coords = np.array([float(v) for v in str(st).split(' ')])[:3]
    temp = np.array([float(v) for v in str(st).split(' ')])[3:]
    if np.var(temp) > vthres:
        #temp = np.sqrt(temp)
        temp = temp - np.mean(temp)
        temp = temp/np.sqrt(np.var(temp))
        return model.predict(temp)
    else:
        return -1
```

```python
labels = rawrdd.map(clas).collect()
```

```python
plt.hist(labels)
```
