# Analysis of separate coil data (chris)

```python
import numpy as np
import numpy.linalg as nla
import matplotlib.pyplot as plt
import nibabel as nib
import os
%matplotlib inline
```

```python
path = '/home/snarles/predator'
```

```python
#os.listdir(path + '/8631_5_1_pfile/coil_images')
```

```python
bvecs = np.loadtxt(path + '/chris1_bvec.csv', delimiter = ' ').T
b0_inds = list(np.where(np.array([nla.norm(v) for v in bvecs]) == 0)[0])
b0_inds = b0_inds[2:]
```

```python
c1 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil6_ec.nii.gz').get_data()
c1 = c1[:, :, :, b0_inds]
c2 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil7_ec.nii.gz').get_data()
c2 = c2[:, :, :, b0_inds]
c3 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil10_ec.nii.gz').get_data()
c3 = c3[:, :, :, b0_inds]
c4 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil11_ec.nii.gz').get_data()
c4 = c4[:, :, :, b0_inds]
c1_mu = np.mean(c1, axis = 3)
c2_mu = np.mean(c2, axis = 3)
c3_mu = np.mean(c3, axis = 3)
c4_mu = np.mean(c4, axis = 3)
```

```python
(np.min(c1_mu[:]), np.max(c1_mu[:])), (np.min(c2_mu[:]), np.max(c2_mu[:])), (np.min(c3_mu[:]), np.max(c3_mu[:]))
```

```python
vmin_ = -2.0
vmax_ = 17.0
figsize_ = (3, 3)
z_inds = [10, 20, 30, 40, 50]
```

## Sep coil images

```python
#w_inds = range(10)
#z_inds = [12, 14, 16, 18, 20]
figsize_2 = (len(z_inds) * figsize_[0], figsize_[1])
f_c1, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c1_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
f_c2, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c2_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
f_c3, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c3_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
f_c4, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c4_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
```

## shared variation
## time is left-right

```python
w_inds = range(10)
vmin_2 = -1.0
vmax_2 = 1.0
figsize_2 = (len(w_inds) * figsize_[1]/2, 4 * figsize_[0])
f_sv, axarr = plt.subplots(4, len(w_inds), sharex=True, sharey=True, figsize = figsize_2)
z_ind = 30
for i in range(len(w_inds)):
    w_ind = w_inds[i]
    axarr[0, i].imshow(c1[:, 60:, z_ind, w_ind] - c1_mu[:, 60:, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
    axarr[1, i].imshow(c2[:, 60:, z_ind, w_ind] - c2_mu[:, 60:, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
    axarr[2, i].imshow(c3[:, 60:, z_ind, w_ind] - c3_mu[:, 60:, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
    axarr[3, i].imshow(c4[:, 60:, z_ind, w_ind] - c4_mu[:, 60:, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
```
