# Analysis of separate coil data (chris)

```python
import numpy as np
import numpy.linalg as nla
import matplotlib.pyplot as plt
import nibabel as nib
import os
%matplotlib inline
```

```python
path = '/home/snarles/predator'
```

```python
#os.listdir(path + '/8631_5_1_pfile/coil_images')
```

```python
bvecs = np.loadtxt(path + '/chris1_bvec.csv', delimiter = ' ').T
b0_inds = list(np.where(np.array([nla.norm(v) for v in bvecs]) == 0)[0])
b0_inds
```

```python
c1 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil19_ec.nii.gz').get_data()
c1 = c1[:, :, :, b0_inds]
c2 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil20_ec.nii.gz').get_data()
c2 = c2[:, :, :, b0_inds]
c3 = nib.load(path + '/8631_5_1_pfile/coil_images/8631_5_coil21_ec.nii.gz').get_data()
c3 = c3[:, :, :, b0_inds]
```

```python
c1_mu = np.mean(c1, axis = 3)
c2_mu = np.mean(c2, axis = 3)
c3_mu = np.mean(c3, axis = 3)
```

```python
(np.min(c1_mu[:]), np.max(c1_mu[:])), (np.min(c2_mu[:]), np.max(c2_mu[:])), (np.min(c3_mu[:]), np.max(c3_mu[:]))
```

```python
vmin_ = -2.0
vmax_ = 17.0
figsize_ = (3, 3)
```

```python
z_inds = [10, 20, 30, 40, 50]
figsize_2 = (len(z_inds) * figsize_[0], figsize_[1])
f_c1_b0, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c1_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
    
f_c2_b0, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c2_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)

f_c3_b0, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c3_mu[:, :, z_ind], cmap='bone', vmin = vmin_, vmax = vmax_)
```

```python
vmin_2 = -2.0
vmax_2 = 2.0
w_ind = 7
#z_inds = [12, 14, 16, 18, 20]
figsize_2 = (len(z_inds) * figsize_[0], figsize_[1])
f_c1, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c1[:, :, z_ind, w_ind] - c1_mu[:, :, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
f_c2, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c2[:, :, z_ind, w_ind] - c2_mu[:, :, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
f_c3, axarr = plt.subplots(1, len(z_inds), sharex=True, sharey=True, figsize = figsize_2)
for j in range(len(z_inds)):
    z_ind = z_inds[j]
    axarr[j].imshow(c3[:, :, z_ind, w_ind] - c3_mu[:, :, z_ind], cmap='bone', vmin = vmin_2, vmax = vmax_2)
```
